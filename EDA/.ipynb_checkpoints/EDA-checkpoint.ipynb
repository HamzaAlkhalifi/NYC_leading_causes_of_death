# New York City Leading Causes of Death

The leading causes of death by sex and ethnicity in New York City in since 2007. Cause of death is derived from the NYC death certificate which is issued for every death that occurs in New York City.

Rates based on small numbers (RSE > 30) as well as aggregate counts less than 5 have been suppressed in downloaded data

Source: [Bureau of Vital Statistics and New York City Department of Health and Mental Hygiene](https://data.cityofnewyork.us/Health/New-York-City-Leading-Causes-of-Death/jb7j-dtam/about_data)

## Columns: 

| Column name | Description |
|-------------|-------------|
| Year | The year of death |
| Leading Cause | The cause of death |
| Sex | The decedent's sex |
| Race Ethnicity | The decedent's ethnicity |
| Deaths | The number of people who died due to cause of death | 

each row represents a Cause of Death

## Imports 

```python
# for data manipulation 
import pandas as pd
import numpy as np 

# for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
df = pd.read_csv("../data/New_York_City_Leading_Causes_of_Death_20251105.csv")
df = df.drop(columns=["Age Adjusted Death Rate", "Death Rate"])
df.head()
```

## Data Cleaning 

1. Dealing with missing data
2. Data formatting and changing data type
3. Removing duplicates if there is any 

```python
print(f"rows: {df.shape[0]} | columns: {df.shape[1]}")
```

```python
# dealing with missing data
print("number of missing values in each column\n")
df.isna().sum()
```

```python
# droping rows where Race Ethnicity are not known and death number is "."
df = df[df["Race Ethnicity"] != "Not Stated/Unknown"]
df = df[df["Deaths"] != "."]
```

```python
print(f"rows: {df.shape[0]} | columns: {df.shape[1]}")
```

```python
df.info()
```

```python
# changing data type
df["Deaths"] = pd.to_numeric(df["Deaths"])
df.tail()
```

```python
df.info()
```

```python
# data formatting Race Ethnicity
df["Race Ethnicity"] = df["Race Ethnicity"].str.replace("Non-Hispanic", "", regex=False).str.strip()
df["Race Ethnicity"] = df["Race Ethnicity"].str.replace("Race/ Ethnicity", "", regex=False).str.strip()
```

```python
# data formatting Leading Cause
df["Leading Cause"] = df["Leading Cause"].str.replace(r"\s*\(.*?\)", "", regex=True)
df["Leading Cause"] = df["Leading Cause"].str.replace("Chronic Liver Diseases and Cirrhosis", "Chronic Liver Disease and Cirrhosis")
df["Leading Cause"] = df["Leading Cause"].str.replace("Accidents Except Drug Posioning", "Accidents Except Drug Poisoning")
df.head()
```

```python
# data formatting Sex
df["Sex"] = df["Sex"].replace("M","Male")
df["Sex"] = df["Sex"].replace("F","Female")
df["Sex"].unique()
```

```python
# removing duplicates 
print(f"number of duplicates: {df.duplicated().sum()}\n")
print("duplicated rows are:")
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
print(f"number of duplicates after droping: {df.duplicated().sum()}\n")
```

## Data Exploration 

1. Summarizing main characteristics of the data
2. Visualizing data
3. Finding patterns

```python
# statistical summmary
df.describe(include="all")
```

```python
# getting the many tiems a causes of deaths is mentioned and its annual avg deaths 
print(f"There is {df["Leading Cause"].nunique()} mentioned causes of deaths")
cause_death = df.groupby(["Year","Leading Cause"]).agg({"Leading Cause":"count", "Deaths":"sum"})
cause_death = cause_death.rename(columns={"Leading Cause":"Count"}).reset_index()
cause_death = cause_death.groupby("Leading Cause").agg({"Count":"sum", "Deaths":"mean"})
cause_death = cause_death.rename(columns={"Deaths":"AVG Deaths"})
cause_death.head(10)
```

```python
# plotting leading causes of death by count of causes & average deaths
fig = plt.figure(figsize=(25,14))

# plot sorted by count
sort_count = cause_death.sort_values("Count", ascending= False).head(10)
ax1 = fig.add_subplot(2,2,1)
ax1_1 = ax1.twinx()

ax1.bar(sort_count.index, height=sort_count.Count, color="#4DA0C9")
ax1.set_title("Causes of Deaths Sorted by Count")
ax1.set_xlabel("Cause of Death")
ax1.tick_params(axis='x', labelsize=10, rotation=90)
ax1.set_ylabel("Count",color="#4DA0C9")
ax1.tick_params(axis='y', labelcolor="#4DA0C9")

ax1_1.plot(sort_count.index, sort_count["AVG Deaths"], linewidth=2, linestyle="-.",marker="o", markersize=5, color="#C94D62")
ax1_1.set_ylabel("Average Deaths",color="#C94D62")
ax1_1.tick_params(axis='y', labelcolor="#C94D62")
ax1_1.grid(color='white', linestyle=':', linewidth=2, alpha=0.5, axis="y")

# plot sorted by AVG. deaths
sort_avg = cause_death.sort_values("AVG Deaths", ascending= False).head(10)
ax1 = fig.add_subplot(2,2,2)
ax1_1 = ax1.twinx()

ax1.bar(sort_avg.index, height=sort_avg.Count, label=sort_avg.index, color="#4DA0C9")
ax1.set_title("Causes of Deaths Sorted by Annual Average Deaths")
ax1.set_xlabel("Cause of Death")
ax1.tick_params(axis='x', labelsize=10, rotation=90)
ax1.set_ylabel("Count",color="#4DA0C9")
ax1.tick_params(axis='y', labelcolor="#4DA0C9")

ax1_1.plot(sort_avg.index, sort_avg["AVG Deaths"], linewidth=2, linestyle="-.",marker="o", markersize=5, color="#C94D62")
ax1_1.set_ylabel("Average Deaths",color="#C94D62")
ax1_1.tick_params(axis='y', labelcolor="#C94D62")
ax1_1.grid(color='white', linestyle=':', linewidth=2, alpha=0.5, axis="y")

fig.suptitle("Leading Causes of Death by Count & Annual Average Deaths")
fig.savefig("../imgs/leading_causes_by_count_and_avg.png")
print("Count means: how many times did the cause got mentioned by each year, race and sex")
plt.show()
```

```python
# grouping by year and getting the total and average deaths by each year
year_death = df.groupby(["Year", "Leading Cause"]).agg({"Deaths":"sum"}).reset_index()
year_death["_"] = year_death["Deaths"]
year_death = year_death.groupby("Year").agg({"Deaths":"sum", "_": "mean"})
year_death.head()
```

```python
# plot the trend line of total deaths and death rate by each year
plt.figure(figsize=(5,3), dpi=150)
plt.title("Annual Total Deaths and Average Deaths")
ax1 = plt.gca()
ax2 = ax1.twinx()

line1, = ax1.plot(year_death.index, year_death.Deaths, color="#764DC9", linewidth=2, label="Total Deaths")
ax1.tick_params(axis="x", labelsize=8) 
ax1.tick_params(axis="y", labelsize=8, labelcolor="#764DC9")
ax1.set_xlabel("Years")
ax1.set_ylabel("Total Deaths", color="#764DC9")

line2, = ax2.plot(year_death.index, year_death._, color="#A0C94D", linewidth=1, label="Avg. Deaths", linestyle="--")
ax2.tick_params(axis="y", labelsize=8, labelcolor="#A0C94D")
ax2.set_ylabel("Average Deaths", color="#A0C94D")

lines = [line1, line2]
labels = [line.get_label() for line in lines]
plt.legend(lines, labels, loc="upper left", fontsize=7, frameon=False)

plt.tight_layout()
plt.savefig("../imgs/annual_total_and_avg_deaths.png")
print("Average of death means: by each Cause of Death")
plt.show()
```

```python
# trying to answer: what are the highest causes of deaths in 2020 and what races died of it
deaths_by_race = df[df["Year"] == 2020].groupby(["Race Ethnicity", "Leading Cause"]).agg({"Deaths":"sum"}).reset_index()
highest_5 = deaths_by_race.groupby("Leading Cause").agg({"Deaths":"sum"}).sort_values("Deaths", ascending=False)
highest_5 = highest_5.head().index.to_list()
deaths_by_race = deaths_by_race[deaths_by_race["Leading Cause"].isin(highest_5)]
```

```python
plt.figure(figsize=(15,6))
plt.title("2019 Total Deaths by Highest 5 Causes")
sns.barplot(data=deaths_by_race, x="Deaths", y="Leading Cause", hue="Race Ethnicity")
plt.grid(color='grey', linestyle='--', linewidth=2, alpha=0.5, axis="x")
plt.savefig("../imgs/2019_total_deaths_highest_5_Causes.png")
plt.show()
```

```python
# Male vs Female
m_vs_f_year = df.groupby(["Year", "Sex"]).agg({"Deaths":"sum"}).reset_index()

m_vs_f_cause = df.groupby(["Year", "Leading Cause", "Sex"]).agg({"Deaths":"sum"}).reset_index()
m_vs_f_cause
```

```python
fig = plt.figure(figsize=(15,6))
fig.suptitle("Male Vs. Female")

plt.subplot(1,2,1)
plt.title("Distribution of Deaths by Leading Cause Over the Years")
sns.kdeplot(data=m_vs_f_cause, x="Deaths", hue="Sex", alpha=0.4, clip=(m_vs_f_cause["Deaths"].min(), m_vs_f_cause["Deaths"].max()))

plt.subplot(1,2,2)
plt.title("Annual Total Deaths")
sns.lineplot(data=m_vs_f_year, x="Year", y="Deaths", hue="Sex", palette="crest")

plt.savefig("../imgs/male_vs_female")
plt.show()
```

```python
# races average deaths by each cause
print("The 50 Highest Average Deaths by Race and Cause")
df.groupby(["Leading Cause", "Race Ethnicity"]).agg(dict(Deaths="mean")).reset_index().sort_values("Deaths",ascending=False).head(50)
```

```python
# deaths by each race and sex over years
ax = sns.lmplot(data=df, x="Year", y="Deaths", row="Sex", col="Race Ethnicity", 
              fit_reg=False, facet_kws=dict(sharex=False, sharey=False))

ax.fig.suptitle("Deaths by Each Race and Sex Over Years", y=1.05)
plt.savefig("../imgs/deaths__race_sex_years.png")
plt.show()
```
